```python
import pandas as pd

# load the global terrorism dataset
df = pd.read_excel('globalterrorismdb_0522dist.xlsx')
```

```python
rqmask = df['country_txt'] == 'Iraq'
smmask = df['country_txt'] == 'Somalia'
symask = df['country_txt'] == 'Syria'

# Combine the masks with an 'or' condition
mask_to_drop = rqmask | smmask | symask

# Use the negation of the combined mask to keep rows where the mask is False
df = df[~mask_to_drop]
```

```python
df=df[df['iyear']>=2000]
```

```python
# Create empty dataframe
new_df = pd.DataFrame(columns=['Perpetrator', 'Attack Type 1', 'Attack Type 2', 'Attack Type 3', 
                               'Weapon Type 1', 'Weapon Type 2', 'Weapon Type 3', 'Target Type 1',
                               'Target Type 2', 'Target Type 3'])

# Create an empty list to store new rows
new_rows_list = []

# Simulate iterating through your global terrorism database 
for index, row in df.iterrows():
    perp_classification = 1 if (row['individual'] == 1 and 
                                row['nperps'] == 1) else 0

    new_row = {'Perpetrator': perp_classification, 
               'Attack Type 1': row['attacktype1'],
               'Weapon Type 1': row['weaptype1'],
               'Target Type 1': row['targtype1'],
               'Attack Type 2': row['attacktype2'], 
               'Weapon Type 2': row['weaptype2'],
               'Target Type 2': row['targtype2'],
               'Attack Type 3': row['attacktype3'], 
               'Weapon Type 3': row['weaptype3'],
               'Target Type 3': row['targtype3'],
               'Weapon Type 4': row['weaptype4']
              }

    # Append the new row to the list
    new_rows_list.append(new_row)

# Concatenate all rows after the loop
new_df = pd.concat([new_df, pd.DataFrame(new_rows_list)], ignore_index=True)

print(new_df)
```

```python
perpetrator_counts = new_df['Perpetrator'].value_counts()
print(perpetrator_counts)
```
