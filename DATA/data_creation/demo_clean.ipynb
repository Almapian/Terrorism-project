```python
# Specify the added feature column names you want to merge from vdem_df
features_to_merge = ['Population']  # Replace these with your actual feature column names

# Adjust the merge to include only the specified added features along with the keys for merging
merged_df = pd.merge(data, demo[['Country name', 'Year'] + features_to_merge],
                     left_on=['country', 'year'],
                     right_on=['Country name', 'Year'],
                     how='left')

merged_df.drop(['Country name', 'Year'], axis=1, inplace=True)
merged_df['Population'].isnull().sum()
```
